```python
#用LSTM學習一個字符到一個字符的映射
#STEP1.匯入Keras及相關模組
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

#跑起來的結果才會是相同的
numpy.random.seed(7)
```

```python
# 定義序列數據集
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# 建立字母映射到整數（0-25)和反向的查詢字典
char_to_int = dict((c, i) for i, c in enumerate(alphabet))
int_to_char = dict((i, c) for i, c in enumerate(alphabet))

# print出來看一下
print("字母對應到的數字編號: \n", char_to_int)
print("\n")
print("數字編號對應到的字母: \n", int_to_char)
```

```python
# 輸入數據集
seq_length = 1
dataX = []
dataY = []
for i in range(0, len(alphabet) - seq_length, 1):
    seq_in = alphabet[i:i + seq_length]
    seq_out = alphabet[i + seq_length]
    dataX.append([char_to_int[char] for char in seq_in])
    dataY.append(char_to_int[seq_out])
    print(seq_in, '->', seq_out)
```

```python
# 把 X 資料的維度變成 (samples, time_steps, features)
X = numpy.reshape(dataX, (len(dataX), seq_length, 1))

# normalize
X = X / float(len(alphabet))

# one-hot 編碼輸出變量
y = to_categorical(dataY)

print("X shape: ", X.shape) # (25筆samples, "1"個時間長, 1個feature)
print("y shape: ", y.shape)
```

```python
# 建立模型
model = Sequential()
model.add(LSTM(32,input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(y.shape[1],activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=500, batch_size=1, verbose=2)
```

```python
# 評估模型的好壞
scores = model.evaluate(X, y, verbose=0)
print("Model Accuracy: %.2f%%" % (scores[1]*100))
```

```python
# 模型預測力
for pattern in dataX:
    # 把26個字母一個個拿進來模型預測會出現的字母
    x = numpy.reshape(pattern, (1, len(pattern), 1))
    x = x / float(len(alphabet))

    prediction = model.predict(x, verbose=0)
    index = numpy.argmax(prediction) # 機率最大的index
    result = int_to_char[index] # 看看預測出來的是哪一個字母
    seq_in = [int_to_char[value] for value in pattern]
    print(seq_in, "->", result)
```

LSTM學習可變長度字符輸入到單字符輸出

```python
# 準備訓練資料
num_inputs = 1000
max_len = 5 # 最大序列長度
dataX = []
dataY = []
for i in range(num_inputs):
    start = numpy.random.randint(len(alphabet)-2)
    end = numpy.random.randint(start, min(start+max_len,len(alphabet)-1))
    sequence_in = alphabet[start:end+1]
    sequence_out = alphabet[end + 1]
    dataX.append([char_to_int[char] for char in sequence_in])
    dataY.append(char_to_int[sequence_out])
    print(sequence_in, '->', sequence_out)
```

```python
# 將訓練資料轉換為陣列和並進行序列填充（如果需要）
X = pad_sequences(dataX, maxlen=max_len, dtype='float32') # <-- 注意這裡

# 重塑 X 資料的維度成為 (samples, time_steps, features)
X = numpy.reshape(X, (X.shape[0], max_len, 1)) # <-- 特別注意這裡

# 歸一化
X = X / float(len(alphabet))

# 使用one hot encode 對Y值進行編碼
y = to_categorical(dataY)
```

```python
# 創建模型
batch_size = 1
model = Sequential()
model.add(LSTM(32, input_shape=(X.shape[1], 1))) # <-- 注意這裡
model.add(Dense(y.shape[1], activation='softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=500, batch_size=batch_size, verbose=2)
```

```python
# 評估模型的性能
scores = model.evaluate(X, y, verbose=0)
print("Model Accuracy: %.2f%%" % (scores[1]*100))
```

```python
# 讓我們擷取1~5個字符轉成張量結構 shape:(1,5,1)來進行infer
for i in range(20):
    pattern_index = numpy.random.randint(len(dataX))
    pattern = dataX[pattern_index]
    x = pad_sequences([pattern], maxlen=max_len, dtype='float32')
    x = numpy.reshape(x, (1, max_len, 1))
    x = x / float(len(alphabet))
    prediction = model.predict(x, verbose=0)
    index = numpy.argmax(prediction)
    result = int_to_char[index]
    seq_in = [int_to_char[value] for value in pattern]
    print(seq_in, "->", result)
```
